```python
import sys
import os
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), '..')))
from extract import LOCM
from utils import read_plan, read_json_file
from evaluator import ExecutabilityEvaluator
from collections import defaultdict
```

```python
domain_traces = defaultdict(list)
raw_file = "../../data/plain_traces/plain_traces.txt"
with open(raw_file, "r") as f:
    lines = f.readlines()
    for line in lines:
        item = line.split("&&")
        domain = item[0].strip()
        trace = read_plan(item[-1])
        domain_traces[domain].append(trace)

def get_stats(TM_list):
    total = 0
    _max = 0
    for TM in TM_list:
        transitions = TM[TM > 0].sum().sum()
        total += transitions
        if transitions > _max:
            _max = transitions
    return total, _max

for domain in domain_traces:
    traces = domain_traces[domain]
    locm = LOCM()
    TM_list = locm.get_TM_list(traces, True)
    stats = get_stats(TM_list)
    print("domain: {}, total transitions: {}, max transitions: {}".format(
        domain, stats[0], stats[1]
    ))

```
